**MELAKUKAN DATA PREPARATION**

```python
#Encoding Categorial Feature
#one-hot-encoding
from sklearn.preprocessing import OneHotEncoder
datas = pd.concat([datas, pd.get_dummies(datas['restaurant'], prefix='restaurant', drop_first=True)], axis=1)
datas = pd.concat([datas, pd.get_dummies(datas['item'], prefix='item', drop_first=True)], axis=1)
datas = pd.concat([datas, pd.get_dummies(datas['salad'], prefix='salad', drop_first=True)], axis=1)
datas.drop(['restaurant','item','salad'], axis=1, inplace=True)
datas.head()
```

```python
#Melakukan Principal Component Analysis
sns.pairplot(datas[['sat_fat','cal_fat','total_fat']], plot_kws={"s": 3});
```

```python
#PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=3, random_state=123)
pca.fit(datas[['sat_fat','cal_fat','total_fat']])
princ_comp = pca.transform(datas[['sat_fat','cal_fat','total_fat']])
pca.explained_variance_ratio_.round(3)
```

```python
#reduksi fitur x, y, z, menjadi fitur ukuran berlian
from sklearn.decomposition import PCA
pca = PCA(n_components=1, random_state=123) #jumlah komponen 1
pca.fit(datas[['sat_fat','cal_fat','total_fat']]) #fit model dengan data masukan
datas['dimension'] = pca.transform(datas.loc[:, ('sat_fat','cal_fat','total_fat')]).flatten() #menambahkan fitur 'dimension'
datas.drop(['sat_fat','cal_fat','total_fat'], axis=1, inplace=True) #drop kolom x, y, z
```

```python
#split dataset menjadi data train dan data test sebelum transformasi (supaya transformasi diterapkan hanya pada data latih)
from sklearn.model_selection import train_test_split

X = datas.drop(["calories"], axis=1)
y = datas["calories"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state=123)
```

```python
print(f'Total # of sample in whole dataset: {len(X)}')
print(f'Total # of sample in whole dataset: {len(X_train)}')
print(f'Total # of sample in whole dataset: {len(X_test)}')
```

```python
#Standardisasi dengan StandardScaler pada data latih
from sklearn.preprocessing import StandardScaler

numerical_features = ['cholesterol', 'sugar', 'dimension']
scaler = StandardScaler()
scaler.fit(X_train[numerical_features])
X_train[numerical_features] = scaler.transform(X_train.loc[:, numerical_features])
X_train[numerical_features].head()
```

```python
#Standarisasi mengubah mean menjadi 0 dan standar deviasi menjadi 1
X_train[numerical_features].describe().round(4)
```
